```python

from sklearn.preprocessing import LabelEncoder
```

```python
data['company_n'] = LabelEncoder().fit_transform(data['company'])
data['job_n'] = LabelEncoder().fit_transform(data['job'])
data['degree_n'] = LabelEncoder().fit_transform(data['degree'])
data_new=data.drop(['company','job','degree'],axis=1)
data_new
```
