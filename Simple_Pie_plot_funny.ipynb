```python
#importing the Matplotlib,the visual artist

import matplotlib.pyplot as plt
```

```python
#importing the Counter which we will wills use the divide the data later on
from collections import Counter
```

```python
#calling on our statistical mate, the cuddly pandas
import pandas as pd
```

```python
#wtf! Stealing corey Shafer's code again!

df=pd.read_csv("https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/02-BarCharts/data.csv")
```

```python
#let's take a look at the life of geeks, oh well no surprise

df
```

```python
#the usual, divinding the datas into two list

ids=df['Responder_id']
languages=df['LanguagesWorkedWith']
languages
```

```python
#I swear i know counter works. We have created it. The i in the loop goes on and on, adds the values in into counter. The split helps on seprating the langugae names

languages_counter=Counter()

for i in languages:
    languages_counter.update(i.split(';'))
```

```python
#We need two lists which act as labels and values for our delicious pie. Used a for loop so that it adds the values into it

language=[]
popularity=[]

for d in languages_counter.most_common(5):
    language.append(d[0])
    popularity.append(d[1])
language
```

```python
#umm the aroma is here let's take a look at it
plt.pie(popularity,labels=language,)
```

```python
#Lets's take a final look at our pie! Hope you liked it. DO let me know if you find it difficult or erm.... funny explode helped python to shine amonst all


plt.style.use('seaborn-talk')

plt.pie(popularity,labels=language,wedgeprops={'edgecolor':'black'},autopct='%1.0f%%',explode=[0,0,0,0.2,0],startangle=90,shadow=True)
plt.title("Programming langauge Popularity Pie Chart")

plt.tight_layout

plt.show
```

```python
#Lets garnish it with something like a frame 
plt.style.use('seaborn-talk')
plt.pie(popularity,labels=language,wedgeprops={'edgecolor':'black'},autopct='%1.0f%%',explode=[0,0,0,0.2,0],startangle=90,shadow=True,counterclock=True,frame=True)
plt.title("Programming langauge Popularity Pie Chart")

plt.tight_layout

plt.show


#Well it does is not helping much Let's go back 
```
