```python
dev_money = df['All_Devs']
py_money = df['Python']
js_money = df['JavaScript']
age=df['Age']
```

```python
plt.style.use('fivethirtyeight')
plt.plot(age,dev_money,label='developers money')
plt.plot(age,py_money,label='python money')
average=58000
plt.fill_between(age,py_money,dev_money,where=py_money>average,interpolate=True,color='red',alpha=0.4,label='above Average')
plt.fill_between(age,py_money,dev_money,where=py_money<average,interpolate=True,color='green',alpha=0.25,label='Below Average')

plt.legend()
plt.title("The salaries of IT pros")
plt.show()
```
