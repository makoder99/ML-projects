```python
table=pd.read_csv("https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/02-BarCharts/data.csv")
table
```

```python
ids=table['Responder_id']
lang_responses = table['LanguagesWorkedWith']
```

```python
language_counter= Counter()
```

```python
for i in lang_responses:
    language_counter.update(i.split(';'))


language=[]
popularity=[]

for d in language_counter.most_common(20):
    language.append(d[0])
    popularity.append(d[1])

print(language_counter)

language.reverse()
popularity.reverse()  
plt.style.use('fivethirtyeight')
plt.barh(language,popularity)
plt.title("Programming languages according its popularity")
plt.xlabel("Number of people")
plt.ylabel("Programming Languages")
plt.tight_layout(rect=[0,0,1.4,1.3])

```
