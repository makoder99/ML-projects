```python
import matplotlib.pyplot as plt
```

```python
minutes = [1, 2, 3, 4, 5, 6, 7, 8, 9]

Ram = [8, 6, 5, 5, 4, 2, 1, 1, 0]
Shyam = [0, 1, 2, 2, 2, 4, 4, 4, 4]
Jodu = [0, 1, 1, 1, 2, 2, 3, 3, 4]
```

```python
labels= ['Ram','Shyam', 'Jodu']
```

```python
plt.style.use('fivethirtyeight')
```

```python
plt.stackplot(minutes,Ram,Shyam,Jodu,labels=labels)

plt.legend(loc='best')

plt.title('Minutes')
plt.tight_layout()
plt.show()
```

```python
plt.stackplot(minutes,Jodu,Shyam,Ram,labels=labels)

plt.legend(loc='best')

plt.title('Minutes')
plt.tight_layout()
plt.show()
```
